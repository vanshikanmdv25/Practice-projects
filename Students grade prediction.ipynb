# Handling missing values

```python
from sklearn.impute import SimpleImputer
```

```python
cat_imputer = SimpleImputer(strategy='most_frequent')
df[df.select_dtypes(include=['object']).columns] = cat_imputer.fit_transform(df.select_dtypes(include=['object']))
```

```python
df.isnull().sum()
```

# Data Visualization

```python
#separating target column and columns not necessary in the analysis from data
X= df.drop(['CGPA','Seat No.'],axis=1)
Y= df['CGPA']
```

```python
fig, axes = plt.subplots(nrows=len(df.columns), ncols=1, figsize=(10, len(df.columns) * 4))

for i, column in enumerate(X):
        sns.countplot(y=column, data=df,ax=axes[i])
```

```python
label_encodings = {}
for col in X:
    df[col], label_encodings[col] = pd.factorize(df[col])
```

```python
print("\nLabel Encodings:")
for col, encoding in label_encodings.items():
    print(f'{col}: {list(encoding)}')
```
